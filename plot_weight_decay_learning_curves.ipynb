```python
import pandas as pd
import matplotlib.pyplot as plt

# Load learning curve data
learning_curve_path = "metadataset/learning_curve_data.csv"
learning_data = pd.read_csv(learning_curve_path)

# Plot the learning curves
plt.figure(figsize=(10, 6))
plt.plot(learning_data["train_size"], learning_data["train_error"], label="Training Error", marker='o')
plt.plot(learning_data["train_size"], learning_data["test_error"], label="Test Error", marker='o')

# Add labels and title
plt.xlabel("Training Set Size (Fraction of Total)")
plt.ylabel("Mean Squared Error")
plt.title("Learning Curves for Weight Decay Prediction Model")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save and show plot
#plt.savefig("metadataset/learning_curve_plot.png")
plt.show()
```
